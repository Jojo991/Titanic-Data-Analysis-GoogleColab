```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Titanic dataset from a CSV file (using the URL provided)
titanic_data = pd.read_csv("https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv")

# 2. Display the first few rows of the dataset to understand its structure
titanic_data.head()

# 3. Clean the Data: Remove rows with missing values
titanic_cleaned = titanic_data.dropna()

# 4. Check for any missing values after cleaning
titanic_cleaned.isnull().sum()

# 5. Analyze the Data:
# Calculate the survival rate by gender
survival_rate_gender = titanic_cleaned.groupby('Sex')['Survived'].mean()

# Calculate the survival rate by passenger class
survival_rate_class = titanic_cleaned.groupby('Pclass')['Survived'].mean()

# Calculate the average age of survivors and non-survivors
average_age = titanic_cleaned.groupby('Survived')['Age'].mean()

# 6. Visualize the Data:
# Plot the survival rate by gender
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_rate_gender.index, y=survival_rate_gender.values)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# Plot the survival rate by passenger class
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_rate_class.index, y=survival_rate_class.values)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Create a histogram comparing the ages of survivors and non-survivors
plt.figure(figsize=(8, 5))
sns.histplot(titanic_cleaned[titanic_cleaned['Survived'] == 1]['Age'], kde=True, color='green', label='Survived', bins=30)
sns.histplot(titanic_cleaned[titanic_cleaned['Survived'] == 0]['Age'], kde=True, color='red', label='Not Survived', bins=30)
plt.title('Age Distribution of Survivors and Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Frequency')  # Fixing the missing closing quotation mark
plt.legend()
plt.show()

# 7. Explore Further Insights:
# Check the column names to confirm if 'Embarked' exists
print("Column names in the dataset:", titanic_cleaned.columns)

# Plot the survival rate by embarkation location if the 'Embarked' column exists
if 'Embarked' in titanic_cleaned.columns:
    survival_rate_embarked = titanic_cleaned.groupby('Embarked')['Survived'].mean()

    plt.figure(figsize=(6, 4))
    sns.barplot(x=survival_rate_embarked.index, y=survival_rate_embarked.values)
    plt.title('Survival Rate by Embarked Location')
    plt.xlabel('Embarked Location')
    plt.ylabel('Survival Rate')
    plt.show()
else:
    print("Column 'Embarked' not found in the dataset!")

# Create age groups and calculate the survival rate for each group
age_groups = pd.cut(titanic_cleaned['Age'], bins=[0, 12, 20, 40, 60, 100], labels=['0-12', '13-20', '21-40', '41-60', '61+'])

survival_rate_age_group = titanic_cleaned.groupby(age_groups)['Survived'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=survival_rate_age_group.index, y=survival_rate_age_group.values)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style="whitegrid")  # This sets the background style to white with gridlines

# Load Titanic dataset from a CSV file (using the URL provided)
titanic_data = pd.read_csv("https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv")

# Clean the Data: Remove rows with missing values
titanic_cleaned = titanic_data.dropna()

# 3. Analyze the Data:
# Calculate the survival rate by gender
survival_rate_gender = titanic_cleaned.groupby('Sex')['Survived'].mean()

# Calculate the survival rate by passenger class
survival_rate_class = titanic_cleaned.groupby('Pclass')['Survived'].mean()

# Calculate the average age of survivors and non-survivors
average_age = titanic_cleaned.groupby('Survived')['Age'].mean()

# 4. Visualize the Data:
# Plot the survival rate by gender
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_rate_gender.index, y=survival_rate_gender.values, palette='Set2')
plt.title('Survival Rate by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()

# Plot the survival rate by passenger class
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_rate_class.index, y=survival_rate_class.values, palette='coolwarm')
plt.title('Survival Rate by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()

# **Age Distribution of Survivors and Non-Survivors** - IMPORTANT PLOT
plt.figure(figsize=(8, 5))
sns.histplot(titanic_cleaned[titanic_cleaned['Survived'] == 1]['Age'], kde=True, color='forestgreen', label='Survived', bins=30)
sns.histplot(titanic_cleaned[titanic_cleaned['Survived'] == 0]['Age'], kde=True, color='tomato', label='Not Survived', bins=30)
plt.title('Age Distribution of Survivors and Non-Survivors', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.show()

# 5. Pairplot for Selected Columns (Age, Pclass, Survived)
sns.pairplot(titanic_cleaned[['Age', 'Pclass', 'Survived']], hue='Survived', palette={0: 'tomato', 1: 'forestgreen'})
plt.suptitle('Pairplot of Age, Pclass, and Survived', fontsize=14, y=1.02)
plt.show()

# 6. Explore Further Insights:
# Check the column names to confirm if 'Embarked' exists
print("Column names in the dataset:", titanic_cleaned.columns)

# Plot the survival rate by embarkation location if the 'Embarked' column exists
if 'Embarked' in titanic_cleaned.columns:
    survival_rate_embarked = titanic_cleaned.groupby('Embarked')['Survived'].mean()

    plt.figure(figsize=(6, 4))
    sns.barplot(x=survival_rate_embarked.index, y=survival_rate_embarked.values, palette='Blues')
    plt.title('Survival Rate by Embarked Location', fontsize=14)
    plt.xlabel('Embarked Location', fontsize=12)
    plt.ylabel('Survival Rate', fontsize=12)
    plt.show()
else:
    print("Column 'Embarked' not found in the dataset!")

# Create age groups and calculate the survival rate for each group
age_groups = pd.cut(titanic_cleaned['Age'], bins=[0, 12, 20, 40, 60, 100], labels=['0-12', '13-20', '21-40', '41-60', '61+'])

survival_rate_age_group = titanic_cleaned.groupby(age_groups)['Survived'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=survival_rate_age_group.index, y=survival_rate_age_group.values, palette='viridis')
plt.title('Survival Rate by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better aesthetics
sns.set(style="whitegrid")

# Load Titanic dataset from a CSV file (using the URL provided)
titanic_data = pd.read_csv("https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv")

# Clean the Data: Remove rows with missing values
titanic_cleaned = titanic_data.dropna()

# 1. Analyze the Data:

# Calculate the survival rate by gender
survival_rate_gender = titanic_cleaned.groupby('Sex')['Survived'].mean()

# Calculate the survival rate by passenger class
survival_rate_class = titanic_cleaned.groupby('Pclass')['Survived'].mean()

# Calculate the average age of survivors and non-survivors
average_age = titanic_cleaned.groupby('Survived')['Age'].mean()

# 2. Visualize the Data:

# **Age Distribution of Survivors and Non-Survivors**
plt.figure(figsize=(8, 5))
sns.histplot(titanic_cleaned[titanic_cleaned['Survived'] == 1]['Age'], kde=True, color='mediumseagreen', label='Survived', bins=30)
sns.histplot(titanic_cleaned[titanic_cleaned['Survived'] == 0]['Age'], kde=True, color='indianred', label='Not Survived', bins=30)
plt.title('Age Distribution of Survivors and Non-Survivors', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.show()

# Plot the survival rate by gender
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_rate_gender.index, y=survival_rate_gender.values, palette='muted')
plt.title('Survival Rate by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()

# Plot the survival rate by passenger class
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_rate_class.index, y=survival_rate_class.values, palette='coolwarm')
plt.title('Survival Rate by Passenger Class', fontsize=14)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()

# 3. Pairplot for Selected Columns (Age, Pclass, Survived)
sns.pairplot(titanic_cleaned[['Age', 'Pclass', 'Survived']], hue='Survived', palette={0: 'indianred', 1: 'mediumseagreen'})
plt.suptitle('Pairplot of Age, Pclass, and Survived', fontsize=14, y=1.02)
plt.show()

# 4. Explore Further Insights:

# Check the column names to confirm if 'Embarked' exists
print("Column names in the dataset:", titanic_cleaned.columns)

# Plot the survival rate by embarkation location if the 'Embarked' column exists
if 'Embarked' in titanic_cleaned.columns:
    survival_rate_embarked = titanic_cleaned.groupby('Embarked')['Survived'].mean()

    plt.figure(figsize=(6, 4))
    sns.barplot(x=survival_rate_embarked.index, y=survival_rate_embarked.values, palette='Blues')
    plt.title('Survival Rate by Embarked Location', fontsize=14)
    plt.xlabel('Embarked Location', fontsize=12)
    plt.ylabel('Survival Rate', fontsize=12)
    plt.show()
else:
    print("Column 'Embarked' not found in the dataset!")

# Create age groups and calculate the survival rate for each group
age_groups = pd.cut(titanic_cleaned['Age'], bins=[0, 12, 20, 40, 60, 100], labels=['0-12', '13-20', '21-40', '41-60', '61+'])

survival_rate_age_group = titanic_cleaned.groupby(age_groups)['Survived'].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=survival_rate_age_group.index, y=survival_rate_age_group.values, palette='viridis')
plt.title('Survival Rate by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()

# 5. Correlation Matrix (Heatmap)
# Select only the numeric columns for correlation analysis
numeric_columns = titanic_cleaned.select_dtypes(include=['float64', 'int64'])  # Only numeric columns

# Compute the Correlation Matrix on the numeric columns
correlation_matrix = numeric_columns.corr()  # This computes the correlation of only numerical columns

# Visualize the Correlation Matrix using a Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f', cbar=True)
plt.title('Correlation Matrix of Titanic Dataset', fontsize=14)
plt.show()
```
